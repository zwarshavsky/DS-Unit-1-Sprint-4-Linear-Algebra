# Vertical Line Test

## 1.1 Create two graphs, one that passes the vertical line test and one that does not.

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Plot Linearly Independent Vectors

# Axis Bounds
plt.xlim(-2,5)          
plt.ylim(-1,3)

# Original Vector
a = [-1.5,.5] 
b = [3, 1]

# Plot Vectors
plt.arrow(1,1, a[0], a[1], linewidth=3, head_width=.05, head_length=0.05, color ='blue')
plt.arrow(0,0, b[0], b[1], linewidth=3, head_width=.05, head_length=0.05, color ='red')

plt.title("This graph does not pass the vertical line test")
plt.show()
```

```
# Plot Linearly Independent Vectors

# Axis Bounds
plt.xlim(-3,4)          
plt.ylim(-1,3)

# Original Vector
a = [-1.5,.5] 
b = [3, 1]

# Plot Vectors
plt.arrow(-0.5,0, a[0], a[1], linewidth=3, head_width=.05, head_length=0.05, color ='blue')
plt.arrow(.5,0, b[0], b[1], linewidth=3, head_width=.05, head_length=0.05, color ='red')

plt.title("This graph passes the vertical line test")
plt.show()
```

## 5.1 In your own words, give an explanation for the intuition behind eigenvalues and eigenvectors.


What about the vector represented by the earth's axis? Does it move? 

We could call this vector the **eigenvector** of the earth's rotation brecause the transformation does not reorient it. Any transformation that you could conceive of possesses eigenvectors - dimensions of space that are not pointed in a new direction by the transformation described by a certain transformation matrix. 

## So what's an eigenvalue?

An eigenvector is any vector that doesn't change its orientation during a given transformation, but an eigenvector might still get scaled by some scalar. It can get stretched or squished, it just can't change the direction in which it is pointing. An **eigenvalue** is the scalar that represents how its corresponding eigenvector was scaled during the transformation. Eigenvectors and eigenvalues always come in pairs.

```
import os.path
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO
import pandas as pd

# Read National Data 
national_url = 'https://www.huduser.gov/portal/datasets/hads/hads2013n_ASCII.zip'
national_file = 'thads2013n.txt'

if os.path.exists(national_file):
    national = pd.read_csv(national_file)
else: 
    z_national = urlopen(national_url)
    zip_national = ZipFile(BytesIO(z_national.read())).extract(national_file)
    national = pd.read_csv(zip_national)
```

```
national1 = national.select_dtypes(include=["float64","int64"])
```

```
mean(national1.T, axis=1)
```

```
#having trouble finding Z

from numpy import array
from numpy import mean
from numpy import std
from numpy import cov
from numpy.linalg import eig


print("Data: \n",national.head(1) )
# calculate the mean of each column
means = mean(national1.T, axis=1)
print("\n Means: \n", means)
# center columns by subtracting column means
centered_data = national1 - means
print("\n Centered Data: \n", centered_data)
# calculate the standard deviation of each column
std_devs = std(national1.T, axis=1)
print('\n Standard Deviations: \n', std_devs)
# complete the standardization by dividing by std_dev
standardized_data = centered_data / std_devs
print("\n Standardized Data: \n", standardized_data)
# calculate covariance matrix of centered matrix
# We're not going to find the covaraicne matrix by the (Z^T)(Z) method, I trust this function more
covariance_matrix = cov(standardized_data.T)
print("\n Covariance Matrix: \n", covariance_matrix)
# eigendecomposition of covariance matrix
values, vectors = eig(covariance_matrix)
print("\n Eigenvectors: \n", vectors)
print("\n Eigenvalues: \n", values)
# project data
P = vectors.T.dot(covariance_matrix.T)
print("\n Projected Data: \n", P.T)
```

```
# Principal Component Analysis
from numpy import array
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# define a matrix
print("Data: \n", national1)
# Standardize the Data
# Instantiate a Standard Scaler object
scaler = StandardScaler()
# Use the object to fit_transform our data
Z = scaler.fit_transform(national1)
print("\n Standardized Data: \n", Z)
# create the PCA instance
# The integer value that we're passing in says 
# how many new columns we want to retain from PCA
pca = PCA(2)
# fit on data
pca.fit(Z)
# access values and vectors
print("\n Eigenvectors: \n", pca.components_)
print("\n Eigenvalues: \n",pca.explained_variance_)
# transform data
B = pca.transform(Z)
print("\n Projected Data: \n", B)
```
